```python
'''
for march 2017 to 2018
for each social
create a bar chart

y-axis: count
x-axis: perc change
'''
```

```python
store -r fb_year_3
```

```python
fb_year_3
fb_year_3.perc_change = fb_year_3.perc_change.astype(float)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

filter_values = np.arange(-75,60,5)
out = pd.cut(fb_year_3['perc_change'], bins=filter_values)
counts = pd.value_counts(out,sort=False)

plt.style.use('seaborn')
plt.figure(figsize=(20, 8))
ax = counts.plot(kind='bar')
plt.xticks(rotation='horizontal')
plt.title('Facebook',loc='left')
for p in ax.patches:
    ax.annotate(np.round(p.get_height()).astype(int), (p.get_x()+p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.show()
```
