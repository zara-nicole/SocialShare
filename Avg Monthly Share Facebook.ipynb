```python
#get the mean of every column in fb_data
fb_avg_month = pd.DataFrame()
avg = []
for column in fb_data:
    avg.append(fb_data[column].mean()*100)

row = pd.Series(avg,fb_data.columns.values)
fb_avg_month = fb_avg_month.append([row],ignore_index=True)
fb_avg_month = fb_avg_month.transpose().reset_index().rename(index=str,columns={'index':'date',0:'avg_share'})
```

```python
#fb_avg_month=fb_avg_month.round(2)
```

```python
import numpy as np
from bokeh.layouts import column
from bokeh.io import show, output_file, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.palettes import Spectral5
from bokeh.transform import factor_cmap
from bokeh.resources import CDN
from bokeh.embed import file_html
from bokeh.models.glyphs import  Text



from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file, output_notebook

output_notebook()
group = fb_avg_month.round(2)

tools = 'pan,wheel_zoom,xbox_select,box_zoom,reset,save'

source = ColumnDataSource(group)

p2 = figure(x_range = fb_avg_month['date'], title="Average % Share Per Month", plot_width=950,plot_height=400,
            toolbar_location='left', tools=tools)

p2.y_range.start = 40
p2.y_range.end = 60
p2.grid.grid_line_alpha = 0
p2.xaxis.axis_label = 'Date'
p2.yaxis.axis_label = '% Share'
p2.ygrid.band_fill_color = "olive"
p2.ygrid.band_fill_alpha = 0.1
p2.title.text_font_size = '13pt'
p2.title.text_color = 'black'

p2.circle('date','avg_share',size=5,color='black', alpha=0.8,source=source)
p2.line('date','avg_share', color='navy', source=source)
p2.add_tools(HoverTool(tooltips=[("Average % Share", "@avg_share")]))
p2.xaxis.visible = True

text_glyph = Text(x='date', y='avg_share', x_offset=10, text='avg_share',text_font_size='8pt',angle=0.4)
text_renderer = p2.add_glyph(source, text_glyph)

#html = file_html(p2,CDN, "monthly average")
show(p2)  # open a browser
```
